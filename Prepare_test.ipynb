```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns 
```

```python
df_test = pd.read_csv("testListings.csv",sep = ',') 
df_test
```

#### Drop Columns to match train set

```python
df_test.drop(['thumbnails', 'longitude','latitude','features','description','title','terraceArea','rooms'], axis=1,inplace = True)
```

#### Analize test set

```python
df_test.isnull().sum()
```

```python
df_test.shape
```

#### Drop columns with few rows

```python
df_test =  df_test.dropna(subset=['locationId', 'bathrooms'], how='any')
```

```python
df_test
```

#### Apply log transformation just like with the training data

```python
# Log transform a single column 
df_test['price'] = np.log(df_test['price'])
df_test['bedrooms'] = np.log(df_test['bedrooms'])
df_test['bathrooms'] = np.log(df_test['bathrooms'])
df_test['totalArea'] = np.log(df_test['totalArea'])
df_test['livingArea'] = np.log(df_test['livingArea'])
df_test['plotArea'] = np.log(df_test['plotArea'])
```

```python
df_test
```

#### Fill empty values just like with training data

```python
df_test['price'].fillna(df_test['price'].median(), inplace=True)
df_test['bedrooms'].fillna(df_test['bedrooms'].median(), inplace=True)
df_test['bathrooms'].fillna(df_test['bathrooms'].median(), inplace=True)
df_test['totalArea'].fillna(df_test['totalArea'].median(), inplace=True)
df_test['livingArea'].fillna(df_test['livingArea'].median(), inplace=True)
df_test['plotArea'].fillna(df_test['plotArea'].median(), inplace=True)
```

```python
df_test
```

```python
df_test.to_csv('final_test_set.csv',index = False)
```
